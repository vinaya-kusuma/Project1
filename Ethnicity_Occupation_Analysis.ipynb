### Impact of automation on jobs by Ethnicity by 2032

### Major Findings:
* Based on the analysis we have found 36% of Whites might be impacted by the automation on the jobs.
* 40% of Blacks or African Americans might be impacted by the automation on the jobs.
* 35% of Asians might be impacted by the automation on the jobs.
* 45% of Hispanic or Latino might be impacted by the automation on the jobs.

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Store file paths in a variable
occupation_data_complete_path = Path("Resources/Occupation_complete.csv")
occupation_data_complete = pd.read_csv(occupation_data_complete_path)
occupation_data_complete
```

```python
ethnicity_occupation_data=occupation_data_complete[['2022 National Employment Matrix code',
       '2022 National Employment Matrix title', 'Employment, 2022', 'Employment, 2032',
       'Employment change, percent, 2022-32', 
        'Typical education needed for entry',
       'Work experience in a related occupation',
       'Automation Probability','Total employed', 'Women',
       'White', 'Black or African American', 'Asian', 'Hispanic or Latino']]

ethnicity_occupation_data=ethnicity_occupation_data[(ethnicity_occupation_data['White'] != '–')] 
ethnicity_occupation_data = ethnicity_occupation_data.dropna(subset=['White', 'Automation Probability'])

ethnicity_occupation_data = ethnicity_occupation_data.sort_values("Automation Probability",ascending=False)
ethnicity_occupation_data
```

```python
def totalWhiteEmployed(row):
    totalEmployedNum =  float(row ['White']) * float(row['Total employed']) /100
    return totalEmployedNum

white_ethnicity_occupation_data = ethnicity_occupation_data[['2022 National Employment Matrix code',
       '2022 National Employment Matrix title', 'Employment, 2022', 'Employment, 2032',
       'Employment change, percent, 2022-32', 
        'Typical education needed for entry',
       'Work experience in a related occupation',
       'Automation Probability','Total employed',
       'White']]

white_ethnicity_occupation_data = white_ethnicity_occupation_data.sort_values("White",ascending=False)
white_ethnicity_occupation_data['Total Employed White'] = white_ethnicity_occupation_data.apply(totalWhiteEmployed, axis=1)
white_ethnicity_occupation_data = white_ethnicity_occupation_data.sort_values("Total Employed White",ascending=False)

TotalEmployedWhite = white_ethnicity_occupation_data['Total Employed White'].sum()
print(TotalEmployedWhite)
PotentialImpactedWhites = white_ethnicity_occupation_data.loc[white_ethnicity_occupation_data['Automation Probability'] > 0.75, 'Total Employed White'].sum()
print(PotentialImpactedWhites)
percentWhiteImpacted = PotentialImpactedWhites *100 /TotalEmployedWhite
print(round(percentWhiteImpacted))

whiteData = [['White', TotalEmployedWhite, PotentialImpactedWhites, percentWhiteImpacted]] 
# Create the pandas DataFrame 
EthinicityAnalysisDF = pd.DataFrame(whiteData, columns=['Ethinicity', 'Total Employed', 'Potential Impacted', '% impacted']) 

white_ethnicity_occupation_data
```

```python
def totalBlackAfricanAmericanEmployed(row):
    totalEmployedNum =  float(row ['Black or African American']) * float(row['Total employed']) /100
    return totalEmployedNum

Black_African_American_ethnicity_occupation_data = ethnicity_occupation_data[['2022 National Employment Matrix code',
      '2022 National Employment Matrix title', 'Employment, 2022', 'Employment, 2032',
       'Employment change, percent, 2022-32', 
        'Typical education needed for entry',
       'Work experience in a related occupation',
       'Automation Probability','Total employed',
       'Black or African American']]

Black_African_American_ethnicity_occupation_data = Black_African_American_ethnicity_occupation_data.sort_values("Black or African American",ascending=False)
Black_African_American_ethnicity_occupation_data['Total Employed Black or African American'] = Black_African_American_ethnicity_occupation_data.apply(totalBlackAfricanAmericanEmployed, axis=1)
Black_African_American_ethnicity_occupation_data = Black_African_American_ethnicity_occupation_data.sort_values("Total Employed Black or African American",ascending=False)


TotalEmployedBlackAfricanAmerican = Black_African_American_ethnicity_occupation_data['Total Employed Black or African American'].sum()
PotentialImpactedBlackAfricanAmerican = Black_African_American_ethnicity_occupation_data.loc[Black_African_American_ethnicity_occupation_data['Automation Probability'] > 0.75, 'Total Employed Black or African American'].sum()
percentBlackAfricanAmericanImpacted = PotentialImpactedBlackAfricanAmerican *100 /TotalEmployedBlackAfricanAmerican


black_row = {'Ethinicity':'Black Or African American', 'Total Employed':TotalEmployedBlackAfricanAmerican, 'Potential Impacted':PotentialImpactedBlackAfricanAmerican, '% impacted':percentBlackAfricanAmericanImpacted}
EthinicityAnalysisDF = pd.concat([EthinicityAnalysisDF, pd.DataFrame([black_row])], ignore_index=True)

Black_African_American_ethnicity_occupation_data
```

```python
white_ethnicity_occupation_data = white_ethnicity_occupation_data.head(10)

white_color_list = []
for index, row in white_ethnicity_occupation_data.iterrows():
    if float(row['Automation Probability']) >= 0.75:
        white_color_list.append('red')
    else:
        white_color_list.append('green')
    
white_bar_plot = white_ethnicity_occupation_data.plot(x="2022 National Employment Matrix title", y="Total Employed White", 
                                                      kind="bar", color=white_color_list, legend=False,
                                                     title='Total White Employed, safe vs unsafe jobs due to automation probability >0.75')
plt.xlabel('2022 National Employment Matrix title')
plt.ylabel('Total Employed White')
```

### Explanation
* We have chosen top 10 jobs whites have been employed to plot the graph. Each bar represents the number of whites employed. 
* The Green Bars shows the jobs where the automation probability is less than 0.75. Means lesser risky jobs.
* The red bars shows the jobs where the automation probability is greater than 0.75. making those risky jobs.
* We computed the potential impacted Whites using total employed vs people employed on risky jobs, calculating to 36% at risk of loosing jobs due to automation.
